```python
import fastai
from fastai.tabular import *
import pandas as pd
```

```python
df = pd.read_csv('../input/marketingrar/marketing.csv')
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
dep_var = 'Visitors'
cat_names = ['Date','Day_Name','Promo']
cont_names = ['Week','Month','Year','Revenue','Marketing Spend']
procs = [FillMissing, Categorify, Normalize]
```

```python
data = (TabularList.from_df(df, cat_names=cat_names, cont_names=cont_names, procs=procs)
       .split_by_idx(valid_idx=())
       .label_from_df(cols=dep_var)
       .databunch())
```

```python
data.show_batch()
```

```python
learn = tabular_learner(data, layers=[200,100], metrics=accuracy)
learn.fit_one_cycle(25, 1e-2)
learn.save('mini_train')
```

```python
learn.show_results()
```
